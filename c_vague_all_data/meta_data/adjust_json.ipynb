```python
import random
def add_object_json(index):
    with open(f"annotations_{index}.json", "r") as f:
        anno = json.load(f)
    files = os.listdir('./')
    txt_files = [_ for _ in files if 'txt' in _]
        
    class MyEncoder(json.JSONEncoder):
        def default(self, obj):
            if isinstance(obj, bytes):
                return str(obj, encoding='utf-8')
            return json.JSONEncoder.default(self, obj)
        
    for txt_file in txt_files:
        if int(txt_file.split("_")[1]) == index:
            # object_id
            obj_id = int(txt_file.split("_")[-1].replace("obj", "").replace('.txt', ''))
            
            mask = np.loadtxt(f"./mask_{index}_obj{obj_id}.txt")
            print(obj_id, len(anno[f'{index}'][1]['objects']), f"./mask_{index}_obj{obj_id}.txt")
            anno[f'{index}'][1]['objects'][obj_id]['mask'] = encode(np.asfortranarray(mask==1))
    src_referent = anno[f'{index}'][1]['objects'][0]
    random.shuffle(anno[f'{index}'][1]['objects'])
    referent_indx = anno[f'{index}'][1]['objects'].index(src_referent)
    
    
    with open(f"annotations_{index}_.json", "w",  encoding="utf-8") as f:
        temp_dict = {"meta_data": anno[f'{index}'][0], "object": anno[f'{index}'][1]['objects'], "utterance": anno[f'{index}'][2]['captions'][0]['caption_true'], "referent": referent_indx}
        json.dump(temp_dict, f, cls=MyEncoder)
        #for element in anno[f'{index}']:
        #    f.write(json.dumps(temp_dict))
        #    f.write('\n')
```

```python
for _ in range(200):
    add_object_json(_)
```
