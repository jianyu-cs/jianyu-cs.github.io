```python
import os
import random
def generate_cancel_rounds(directory="blue_sphere_and_blue_cube", index=0,count=100):
    behind, behind_mask, front, front_mask, left, left_mask, right, right_mask, no_images, no_image_masks=sample_upperward_rounds(directory)
    start = 0
    #index = 0 
    while start < count:
        
        try:
            sample_round_upperward_cancel_copy(front, front_mask, no_images, no_image_masks, direction="front", directory=directory, index=index+start)
        except:
            print(1)
            return None
            continue
        start += 1
        
        try:
            sample_round_upperward_cancel_copy(right, right_mask, no_images, no_image_masks, direction="right", directory=directory, index=index+start)
        except:
            print(2)
            continue
        start += 1
        
        try:
            sample_round_upperward_cancel_copy(left, left_mask, no_images, no_image_masks, direction="left", directory=directory, index=index+start)
        except:
            print(3)
            continue
        start += 1
        
        try:
            sample_round_upperward_cancel_copy(behind, behind_mask, no_images, no_image_masks, direction="behind", directory=directory, index=index+start)
        except:
            print(4)
            continue
        start += 1
    return start+index
```

```python
import os
import random
def generate_nonmonotonic_rounds(directory="blue_sphere_and_blue_cube", index=0,count=100):
    behind, behind_mask, front, front_mask, left, left_mask, right, right_mask, no_images, no_image_masks=sample_nonmonotonic_rounds(directory)
    start = 0
    #index = 0 
    while start < count:
        
        try:
            sample_round_cancel_nonmonotonic_copy(front, front_mask, no_images, no_image_masks, direction="front", directory=directory, index=index+start)
        except:
            print(1)
            return None
            continue
        start += 1
        
        try:
            sample_round_cancel_nonmonotonic_copy(right, right_mask,no_images, no_image_masks, direction="right", directory=directory, index=index+start)
        except:
            print(2)
            continue
        start += 1
        
        try:
            sample_round_cancel_nonmonotonic_copy(left, left_mask, no_images, no_image_masks, direction="left", directory=directory, index=index+start)
        except:
            print(3)
            continue
        start += 1
        
        try:
            sample_round_cancel_nonmonotonic_copy(behind, behind_mask, no_images, no_image_masks, direction="behind", directory=directory, index=index+start)
        except:
            print(4)
            continue
        start += 1
    return start+index
```
